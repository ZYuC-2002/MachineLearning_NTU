# 簡單線性回歸 Simple Linear Regression

## 資料匯入

```python
import pandas as pd

url = "https://raw.githubusercontent.com/GrandmaCan/ML/main/Resgression/Salary_Data.csv"

data = pd.read_csv(url)

data
```

```python
# y = w * x + b
# 看薪水跟年資的關係 => x:YearsExperience / y:Salary

x = data["YearsExperience"]  # column "YearsExperience"
x
y = data["Salary"]  # column "Salary"
y
```

## 畫圖

```python
import matplotlib.pyplot as plt

plt.scatter(x, y)  # 散布圖

def y_predict(w, b):
    y = w * x + b
    plt.plot(y)  # 折線圖

y_predict(1, 1)

plt.show()

# y_predict要越接近點點越好
```
